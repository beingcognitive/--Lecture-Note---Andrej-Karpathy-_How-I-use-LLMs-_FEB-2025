# 16. Image input, OCR

이제 다음 mode로 이미지를 살펴보자. 

우리가 위에서 audio chunk를 text token과 같은 관점으로 바라보고, 동일한 1차원 시퀀스로 취급해서 Transformer 구조로 다음에 올 audio token을 예측하게 만들었다면,

이미지의 경우도 image patch로 잘게 짤라서 동일한 1차원 시퀀스로 취급하여 Transformer 구조로 다음에 올 image patch를 예측하게 만들 수 있음. (간단한 방식으로는, 이미지를 grid형태의 patch로 잘라내고, 각 patch를 예를 들자면 약10만개의 patch vocabulary 중 가장 유사한 것으로 매칭하여 1차원 시퀀스로 처리하는 방식)


<b>여기서, Transformer 아키텍처의 정말 놀라운 점은 토큰시퀀스가 텍스트, 오디오, 이미지와 같은 어떤 형태이건 무관하게 동작한다는 점이다. (* modality-agnostic...?)</b>

모델 앞뒤로 Encoder와 Decoder에서 입출력되는 토큰들이 텍스트/오디오/이미지 중 무엇인지는 모델과 별도로 처리하면 되고, (* modality-specific preprocessing)

Transformer 모델은 next token prediction 을 통해 그 통계적인 패턴을 읽어내는 것을 잘 하기만 하면 될 뿐이다...!

# 17. Image output, DALL-E, Ideogram, etc.

이미지 -> 텍스트 과정을 뒤집는다면, 텍스트 -> 이미지를 생성하게 만들 수 있음.

Andrej는 이미지 생성 기능을 자주 쓸 일은 없는 편. YouTube 썸네일 만들 때에 Ideogram, ChatGPT를 번갈아가면서 사용하는 정도.

참고로, ChatGPT에서 이미지 생성되었다고 채팅창에 나오기는 하지만, 사실 Dall-E로 인한 이미지 생성은 별도 프로그램으로 진행되고, ChatGPT는 이미지 생성에 필요한 프롬프트까지만 대화창에 들고 있음.
