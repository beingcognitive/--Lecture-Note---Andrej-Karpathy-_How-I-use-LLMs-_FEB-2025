# 22. Summary

우선, 각 모델별로 약간씩의 특징들을 갖고 있음.

. ChatGPT는 가장 선도적이고 다양한 기능을 보유하고 있음. 디폴트 사용이라고 볼 수 있음.

. ChatGPT에 검색 feature가 도입되었지만, 아직 검색은 Perplexity가 잘 한다고 생각함.

. Claude는 간단한 웹앱 프로토타입 개발 혹은 다이어그램 시각화을 위해 Artifacts가 유용함. (아직 ChatGPT에 없는 기능)

. 모델과 대화할 수 있는 True Audio기능을 사용하기 위해 ChatGPT의 Advanced Voice Mode가 상당히 괜찮음. 단, 제약이 많아서 refusal이 많은 편이기 때문에, 엔터테인먼트적인 측면에서는 Grok을 사용.
