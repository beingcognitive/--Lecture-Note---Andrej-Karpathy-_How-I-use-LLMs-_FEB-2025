# 6. Tool use: internet search

위에서 살펴본 Step 1,2,3를 거친 모델들에게 그 자체의 지식만으로 동작하는 것을 넘어서, 유용한 tool을 사용할 수 있게 해보자. 특히, 가장 유용한 것은 Internet Search툴임.

예를 들면, 최근에 Andrej Karpathy는 최근 White Lotus라는 TV쇼를 재미있게 봤다고 합니다. 다음 에피소드가 언제 나오는지 궁금해졌음.

그런데, 모델의 자체 지식은 대부분 6개월 이전의 데이터를 기반으로 학습을 마쳤기 때문에, 이와 같은 최신 정보를 알 수 없음.

이를 사람이 직접 한다면, 구글링을 하면서, 여러 웹페이지를 돌아다니면서, <b>"광고를 피하면서"</b>, 얻어진 정보를 종합하여 정보를 얻어내야 함.

==> 이를 대신해 LLM에게 검색을 시키면,

LLM이 1차원 토큰 시퀀스에 답변을 적다가 Search The Internet에 해당하는 토큰을 뱉어내고 나서 잠시 토큰 생성 _sampling을 멈추고 검색을 시작함.

그 다음에, 이렇게 검색으로부터 얻어진 본문의 내용들을 다시 토큰 시퀀스에 추가를 시킴.

이 토큰 시퀀스 _context window는 LLM이 직접 참조할 수 있는 working memory이기 때문에,

knowledge cutoff로 인해 모델 자체가 몰랐던 내용들도 검색을 통해서 얻어진 내용으로 우리가 원하는 적절한 답변을 할 수 있게 됨.






Perplexity가 이 방향으로 가장 먼저 확신을 갖고 Search 가 가능한 AI를 만들어냈음. 검색 과정에서 얻어진 문서들에 대해 출처_Citation까지 확인 가능 (* 참고로, Perplexity.ai CEO인 Aravind Srinivas는 예전 Lex 논문작성 예시 등에서 Citation을 통한 신뢰도 상승과 및 연관 질문을 통한 호기심 자극 등의 UI/UX에 신경을 굉장히 많이 썼다고 Fridman 팟캐스트에서 밝힌 바 있습니다.)
