# 12. Cursor: Composer, writing code



잠깐 얘기를 틀어서, 코딩 툴 Cursor 얘기를 해보자.

업무상 코딩을 많이 하는 사람 입장에서는, ChatGPT 웹페이지에 들어가서 매번 내가 작업하는 내용에 대해서 전체 context를 전달하면서 코딩의 도움을 받는 것은 번거로운 일임.

이에, Cursor와 같은 별도 어플리케이션을 통해서 (* 개발자들이 사용하는 Vscode와 같은 IDE_개발 환경의 일종으로, 오픈소스인 Vscode를 포크하여 AI코딩이 용이하도록 새롭게 만들었음.)



약간 디테일한 얘기이지만, 

(1) Cursor 초창기에는 Ctrl-K 를 눌러서, "이 부분을 이렇게 고쳐줘"라고 시작했었고,



(2) 그 다음에는 Ctrl-L을 눌러서, "이 코드를 해석해줘"로 발전했고,



(3) 지금은, Ctrl-I를 눌러서, Composer 기능을 사용하면, 최근 Agent 동작 방향과 같이, 요구사항을 러프하게 적으면, AI가 알아서 직접 코딩을 다 수행하는 단계에 이르렀음. 

    Andrej는 이런 최근 개발 트렌드를 "Vibe Coding"이라고 네이밍하였고, (* 우리말로는 입코딩보다는 긍정적인 느낌으로, 느낌적인 느낌만 담긴 요구사항으로도 알아서 잘 코딩하는 "알잘딱깔센 코딩" 정도로 번역되지 않을까 생각합니다.)

    예를 들어, Andrej가 Tic-Tac-Toe 앱을 만든다고 할 때에, 아래와 같이 "게임에서 이겼을 때에 confetti (색종이를 흩뿌리는 효과) 같은 것을 넣어줘-"라는 요구사항만 Composer에 넣으면,

    스스로 필요한 패키지 설치를 사용자 컨펌 하에 진행하고, 



    자발적으로 '이왕이면 사운드 효과도 넣어야겠다'라고 스스로 판단해서, mp3파일을 다운로드 받아옴



그 결과, confetti 효과 및 사운드 효과 모두 한번에 성공시킴... :)



그래서, Composer 기능을 통한 Vibe Coding _알잘딱깔센 코딩 으로도 상당히 좋은 결과를 낼 수 있는 것을 확인할 수 있었고,

설혹 모든게 잘못되더라도, 그럼 코드를 직접 살펴보면서 수정하면 됨.
